```python
import re#импортируем модуль регулярных выражений  
import numpy
 

#открытие файла для чтения с помощью встроенной функции open
file_obj = open('sentences.txt', 'r')

content = file_obj.read()#сохраняем содержимое файла в строку
content = content.lower()#переводим строку в нижний регистр
content = re.split('[^a-z]', content)  #делаем токенизацию, т.е. разделяем на слова, большой список всех слов
#content = re.sub('', repl, content) 




content=filter(lambda x: len(x)>0,content)#фильтруем наш список. в результате остаются элементы, длина которых больше нуля,т.е. отсеиваются пустые строки

content = list (set(content))#удаляем дубли c помощью трансформации списка в множество, а потом обратно в список

#content = {content: range (len(content)} 

number_slov = len(content)#считаем количество слов 

#content = dict(zip(range (number_slov), content))#преобразовываем наш список в словарь  

file_obj.close()

#print content




##########################################################################################
#####заново открываем файл, чтобы посчитать количество строк в нём для последующего создания матрицы
file_obj = open('sentences.txt', 'r')
content_save = file_obj.readlines()#сохраняем содержимое файла в список, чтобы посчитать количество строк


#content_save[0].count('dogs')#для ПРИМЕРА подсчитаем вхождение слова догс в первый элемент списка, т.е. в первое предложение

number_predloghenij = len (content_save)
file_obj.close()
################################################################################
#################################################################################



nash_array = numpy.zeros((number_predloghenij,number_slov), dtype = int)#создаем матрицу с нулевыми значениями, потом будет некоторые нули заполнять данными, остальные останутся нулями



#nash_array[0][253]#выведем для примера последний элемент первой, она же нулевая строка матрицы


index_i = 0 #считаем итерации i
for i in content_save: #i - это строка из списка предложений. т.е. первое предложение и т.д.
     
     i = i.lower()
     index_x = 0 #считаем итерации x
     for x in content: #x - это слова из списка слов
         
         nash_array [index_i][index_x] = i.count(x) #записываем в соответствующую ячейку матрицы количество вхождений слова x в предложение i
         index_x += 1    
     index_i += 1 
    
 
#nash_array [2] тестовый вывод строки
#nash_array.shape #проверяем размер массива


from scipy.spatial import distance

cosine_result = []#создаем словарь, куда будет записывать результаты просчёта
 
for row in nash_array:
    
    cosine_result.append(distance.cosine(nash_array[0,:], row))#считаем косинусное расстояние от нулевой первой строки до остальных, результат каждой итерации автоматически методом append автоматически добавляется в конец списка
     
    
cosine_result 

from heapq import nsmallest#импортируем интересную функцию, которая выдаёт минимальные значения из списка

#по заданию нам необходимо получить два минимальных косинусных расстояния от нулевой строки, но так как самое минимальное расстояние будет нуль, т.е. косинусное расстояние до нулевой строки, то получаем не два, а ТРИ значения, соответственно перввое будет нулевым, а два других как раз нужные нам
a, b, c = map(cosine_result.index, nsmallest(3, cosine_result))#три минимальных значения мы записали в три произвольных переменных a,b,c
 
print cosine_result    
    
a,b,c   
    
```
